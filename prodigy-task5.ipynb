```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\KABEERJR\Desktop\internship program\sample\US_Accidents_March23_sampled_500k.csv")
data.head(20)
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data['Start_Time'].str.len().value_counts()
```

```python
data=data[data['Start_Time'].str.len()==19]
data=data.dropna(subset=['City'])
```

```python
data.info
```

```python
data['Start_Time']=pd.to_datetime(data['Start_Time'], format='%Y-%m-%d %H:%M:%S')
data['Start_date']=data['Start_Time'].dt.date
data['Start_Time']=data['Start_Time'].dt.time

data['Start_Time']=data['Start_Time'].astype(str).str[:2]
data['Start_Time']=data['Start_Time'].astype(int)
data['Start_Time'].value_counts()
                                  
```

```python
data.describe()
```

CORRELATION MATRIX TO UNDERSTAND THE CORELATED-DATA 

```python
numerical_cols = data.select_dtypes(include='number')

correlation_matrix = numerical_cols.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

VISUALIZATION OF ACCIDENTS BASED ON TIME OF THE DAY,DAYS,MONTHS,YEARS

```python
# Visualize distribution of accidents over time of day
plt.figure(figsize=(10, 6))
sns.countplot(x='Start_Time', data=data)
plt.title('Distribution of Accidents by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
data['Start_date'] = pd.to_datetime(data['Start_date'])
data['Day_of_Week'] = data['Start_date'].dt.day_name()

plt.figure(figsize=(10, 6))
sns.countplot(x='Day_of_Week', data=data, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Distribution of Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
data['Start_date'] = pd.to_datetime(data['Start_date'])
data['Month'] = data['Start_date'].dt.month_name()

plt.figure(figsize=(10, 6))
sns.countplot(x='Month', data=data, order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
plt.title('Distribution of Accidents by Month of the Year')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
data['Start_date'] = pd.to_datetime(data['Start_date'])
data['Year'] = data['Start_date'].dt.year

# Step 4: Plot the distribution of accidents over years
plt.figure(figsize=(10, 6))
sns.countplot(x='Year', data=data)
plt.title('Distribution of Accidents by Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

VISUALIZATION OF ACCIDENTS BASED ON STATES

```python
plt.figure(figsize=(14, 6))
sns.countplot(x='State', data=data, order=data['State'].value_counts().index)
plt.title('Distribution of Accidents by State')
plt.xlabel('State')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

MAJOR CITIES IN CALAFORNIA STATE ACCIDENTS

```python
CA_df=data[data['State']=='CA']
major_cities=CA_df['City'].value_counts()[:15]
sns.barplot(x=major_cities.index,y=major_cities.values)
plt.xticks(rotation=90)
```


 
 
 Visualize distribution of accidents over weather conditions,wind speed,humidity,pressure,precipitation,visiblity



 

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Weather_Condition', data=data, order=data['Weather_Condition'].value_counts().index[:10])
plt.title('Distribution of Accidents by Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data['Humidity(%)'], fill=True)
plt.title('Distribution of Accidents by Humidity')
plt.xlabel('Humidity (%)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data['Pressure(in)'], fill=True)
plt.title('Distribution of Accidents by Pressure')
plt.xlabel('Pressure (inches)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data['Wind_Speed(mph)'], fill=True)
plt.title('Distribution of Accidents by Wind Speed')
plt.xlabel('Wind Speed (mph)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Visibility(mi)', data=data, order=data['Visibility(mi)'].value_counts().index[:10])
plt.title('Distribution of Accidents by Visibility')
plt.xlabel('Visibility (miles)')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data['Precipitation(in)'], fill=True)
plt.title('Distribution of Accidents by Precipitation')
plt.xlabel('Precipitation (inches)')
plt.ylabel('Density')
plt.show()
```
